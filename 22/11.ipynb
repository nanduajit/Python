```python
import pandas as pd
df= pd.read_csv("cups of coffee vs hours of sleep.csv")
import numpy as np
import plotly.express as px
figure=px.bar(df,x="Coffee in ml",y="sleep in hours",title="cups of coffee vs hours of sleep")
print("Correlation factor:",np.corrcoef(df['Coffee in ml'],df['sleep in hours']))
figure.show()
```

It is been observed that higher the amount of intake of coffee, lesser will be  the number of hours of sleep. After calculating the correlation we can understand that the amount of intake of coffee and Sleep Hours are inversely correlated
